## Subgroup

```python
subgroup_list = []
```

```python
def subgroup(num):
  sdf = df.query('Group ==' + str (num))
  print(sdf)
  return sdf
```

```python
for i in range(0, 101, 10):
  subgroup_list.append(subgroup(i))
```

```python
subgroup_list[0] 
```

```python
def subgroup_coounts(num):
  subgroup_list[num] 
  counts = subgroup_list[num].Match.value_counts()
 
  print(percentage(counts[1], subgroup_list[0].shape[0]))
  return percentage(counts[1], subgroup_list[0].shape[0])  
```

```python
sub_df = pd.DataFrame()
sub_df['Subset'] = ['0:100', '10:90', '20:80', '30:70', '40:60', '50:50', '60:40', '70:30', '80:20', '90:10', '100:0']
sub_df
```

```python
sub_sta = []
for i in range(11):
  #subgroup_coounts(i) 
  #float(subgroup_coounts(i).replace("%", ""))
  
  sub_sta.append(subgroup_coounts(i))
```

```python
sub_df['Subset_statistics'] = sub_sta

filepath = "/content/drive/My Drive/creative_cycleland/facade/gradient/5000_analysis_result_sectional.csv"

sub_df.to_csv(filepath)
```
